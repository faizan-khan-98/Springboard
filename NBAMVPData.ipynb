NBA MVP Historical Data

Things to consider:
1. Players with most MVPs
2. Heights of MVPs
3. Heights of MVPs by decade
4. Positions of MVPs
5. Positions of MVPs by decade
6. Teams with most MVPs
7. Teams with most MVPs by decade
8. MVP status in playoffs
9. Most team championships
10. Most team championships over time (time series line chart)

```python
import pandas as pd
import matplotlib.pyplot as plt
mvps = pd.read_excel('NBA_Finals.xlsx')
mvps.head()
```

```python
#1 Players with most MVPs
fig, ax = plt.subplots()
mvps['MVP Name'].value_counts().plot(ax=ax, kind='bar')
fig.set_figwidth(10)
```

```python
#2
mvps['MVP Height (ft)'].hist()
```

```python
#3
import numpy as np
conditions = [(mvps['Year'] < 1960),
             (mvps['Year'] < 1970)& (mvps['Year'] > 1959),
             (mvps['Year'] < 1980)& (mvps['Year'] > 1969),
             (mvps['Year'] < 1990)& (mvps['Year'] > 1979),
             (mvps['Year'] < 2000)& (mvps['Year'] > 1989),
             (mvps['Year'] < 2010)& (mvps['Year'] > 1999),
             (mvps['Year'] < 2020)& (mvps['Year'] > 2009)]
values = ['1950s','1960s','1970s','1980s','1990s','2000s','2010s']
mvps['Decade'] = np.select(conditions,values)

mvps['MVP Height (ft)'].hist(by=mvps['Decade'],figsize=(10,10))
```

```python
#4
mvps['MVP Position'].hist()
```

```python
#5
mvps['MVP Position'].hist(by=mvps['Decade'],figsize=(10,10))
```

```python
#6
mvps = mvps.replace({'Minneapolis': 'Los Angeles'}, regex=True)
mvps = mvps.replace({'Philadelphia Warriors': 'Golden State Warriors'}, regex=True)
mvps = mvps.replace({'Fort Wayne': 'Detroit'}, regex=True)
mvps = mvps.replace({'St. Louis Hawks': 'Atlanta Hawks'}, regex=True)
mvps = mvps.replace({'San Francisco': 'Golden State'}, regex=True)

fig, ax = plt.subplots()
mvps['MVP Team'].value_counts().plot(ax=ax, kind='bar')
fig.set_figwidth(10)
```

```python
#7
mvps['MVP Team'].hist(by=mvps['Decade'],figsize=(15,20))
```

```python
#8
fig, ax = plt.subplots()
mvps['MVP status'].value_counts().plot(ax=ax, kind='bar')
fig.set_figwidth(10)

mvps['MVP status'].hist(mvps['Year'] > 2000,figsize=(10,10))
```

```python
#9
fig, ax = plt.subplots()
mvps['NBA Champion'].value_counts().plot(ax=ax, kind='bar')
fig.set_figwidth(10)
```

```python
#10
mvps['NBA Champion'].hist(by=mvps['Decade'],figsize=(20,20))
```
