Total Passengers at each airlines at SFO from 2006-2015

```python
byairline = airtraffic.groupby(['Published Airline']).sum()
byairline = byairline.drop(columns=['Year'])
byairline = byairline.sort_values(by='Passenger Count',ascending = False)
print(byairline)
```

```python
byairline.reset_index(inplace=True)
```

```python
from __future__ import division

growth = []
current = []
for x in byairline['Published Airline']:
        if ((yearlydata.index.isin([(2015, x)]).any() == True) & (yearlydata.index.isin([(2006, x)]).any() == True)):
            growth.append(((((yearlydata.loc(axis=0)[2015, x])
            - ((yearlydata.loc(axis=0)[2006, x]))))
            / (yearlydata.loc(axis=0)[2006, x])).iloc[0])
            current.append(yearlydata.loc(axis=0)[2015, x].iloc[0])
        elif ((yearlydata.index.isin([(2015, x)]).any() == True) & (yearlydata.index.isin([(2007, x)]).any() == True)):
            growth.append(((((yearlydata.loc(axis=0)[2015, x])
            - ((yearlydata.loc(axis=0)[2007, x]))))
            / ((yearlydata.loc(axis=0)[2007, x]))).iloc[0])
            current.append(yearlydata.loc(axis=0)[2015, x].iloc[0])
        elif ((yearlydata.index.isin([(2015, x)]).any() == True) & (yearlydata.index.isin([(2008, x)]).any() == True)):
            growth.append(((((yearlydata.loc(axis=0)[2015, x])
            - ((yearlydata.loc(axis=0)[2008, x]))))
            / ((yearlydata.loc(axis=0)[2008, x]))).iloc[0])
            current.append(yearlydata.loc(axis=0)[2015, x].iloc[0])
        else:    
            growth.append(0)
            current.append(0)
                      
```

```python
byairline['Growth'] = growth
byairline['2015_Count'] = current
byairline.drop(columns=['level_0','index'],inplace=True)
```

```python
Fastest Growing Airlines at SFO:
```
